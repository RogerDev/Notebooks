```python
from because.probability import prob

from because.synth import read_data
from because.visualization import viz
```

```python
r = read_data.Reader('/home/roger/Datasets/weight-height.csv')
ds = r.read()
ps = prob.ProbSpace(ds)
print('ps.N = ', ps.N)
print('ps.fieldList = ', ps.fieldList)
```

```python
ps.E('Height')
```

```python
ps.E('Weight')
```

```python
d= ps.distr('Gender')
print(d.ToHistTuple())
```

```python
ps.P(('Gender','Male'), ('Height', 72, None))
```

```python
ps.P(('Gender', 'Male'))
```

```python
viz.show(probspace=ps, targetSpec=['Height'], condSpec=[], gtype='pdf', power=1, enhance=True)
viz.show(probspace=ps, targetSpec=['Height'], condSpec=[], filtSpec = [('Gender', 'Male')], gtype='pdf', power=1, enhance=True)
viz.show(probspace=ps, targetSpec=['Height'], condSpec=[], filtSpec = [('Gender', 'Female')], gtype='pdf', power=1, enhance=True)
viz.show(probspace=ps, targetSpec=['Weight'], condSpec=[], gtype='pdf', power=1, enhance=True)
viz.show(probspace=ps, targetSpec=['Weight'], condSpec=[], filtSpec = [('Gender', 'Male')], gtype='pdf', power=1, enhance=True)
viz.show(probspace=ps, targetSpec=['Weight'], condSpec=[], filtSpec = [('Gender', 'Female')], gtype='pdf', power=1, enhance=True)

```

```python
viz.show(probspace=ps, targetSpec=['Height'], condSpec=['Weight'], gtype='pdf', power=1, enhance=True)
```

```python
print(ps.P(('Gender','Male'), [('Height', 66), ('Weight', 180)]),
ps.P(('Gender','Female'), [('Height', 62), ('Weight', 115)]))

```
